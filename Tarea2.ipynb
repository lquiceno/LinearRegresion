De los estimados obtenidos para los parámetros de cada instancia de regresión, calcule el promedio y la desviación estándar, además grafique sus datos en un histograma y un diagrama de cajas.
Discuta sus resultados: el valor del promedio y la desviación estándar, la forma de la distribución resultante...

```python
#modelo inicial
x = 2*np.random.rand(100,1) 

y = 4 +3*x + 2*np.random.rand(100,1)
```

```python
plt.style.use("seaborn")
plt.plot(x,y, 'o')
plt.title('Datos iniciales')
plt.xlabel('x')
plt.ylabel('y');
```

```python
#haciendo la regresión

def train(x,y): #función que hace regresión
    linear_regression = LinearRegression() #instancia el objeto de la calse 
    model = linear_regression.fit(x,y) 
    return model

model = train(x,y) #llama el modelo 

x0 = model.intercept_

x1 = model.coef_

print(f'El valor de los parámetros optimo son {x0} y {x1}')
```

```python
x0 = [] #contiene los valores x0 o intercepto
x1 = [] #contiene los valores x1 o pendiente

for i in range(1000):
    x = 2*np.random.rand(100,1) 
    y = 4 +3*x + 2*np.random.rand(100,1)
    model = train(x,y)
    x0.append(model.intercept_)
    x1.append(model.coef_)
    
```

```python
print(f'El promedio de x0 o interceptos son: {(np.sum(x0))/len(x0)}. El promedio de x1 o pedientes es: {(np.sum(x1))/len(x1)}')
```

Observamos que al realizar mil corridas con datos aleatorios diferentes en cada paso se obtiene medas cercanas a las obtenidas con una sola iteración, aún así, se puede ver la diferencia de estos valores.

```python
print(f'La media de los x0 es: {np.mean(x0)}. La media de los x1 es: {np.mean(x1)}')
```

```python
#saca media 
plt.style.use("seaborn")
sns.distplot(x0).set(title='Histograma con media de x0');
```

```python
plt.style.use("seaborn")
sns.distplot(x1).set(title='Histograma con media de x1'); 
```

Los datos se distribuyen de forma uniforme alrededor de de sus medias.

```python
#Diagramas de cajas 
plt.style.use("seaborn")
sns.boxplot(x0).set(title='Histograma con media de x0'); 
```

```python
plt.style.use("seaborn")
sns.boxplot(x1).set(title='Histograma con media de x1'); 
```

En el diagrama de cajas (es una vista superior de los datos) se lograr ver que hay muy pocos puntos por fuera de la la desviación estandar, en el primer diagrma de cajas los datos esta en el promedio de su desviación estandar de 5 y en el segundo diagrama de caja se distribuyen sobre 3.
